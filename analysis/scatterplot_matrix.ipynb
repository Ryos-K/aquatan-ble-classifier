# データ分析

データを散布図行列にプロットする．

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DIRECTORY = "../formatted"

time_windows = [30, 45, 60]

for time_window in time_windows:
    # データの読み込み
    df_8302 = pd.read_csv(f"{DIRECTORY}/8302_t={time_window}.csv")
    df_8303 = pd.read_csv(f"{DIRECTORY}/8303_t={time_window}.csv")

    # データをランダムにサンプリング
    df_8302 = df_8302.sample(n=100)
    df_8303 = df_8303.sample(n=100)

    #　データを結合
    df = pd.concat([df_8302, df_8303])

    # データのシャッフル
    df = df.sample(frac=1).reset_index(drop=True)

    # 散布図行列のプロット
    pg = sns.pairplot(df, hue="label", palette={"8-302": "orange", "8-303": "deepskyblue"}, height=1.5, plot_kws={"s": 3})
    pg.figure.suptitle(f"t={time_window}", y=1.01)

    
```
